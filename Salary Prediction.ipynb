```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('Salary_dataset.csv')
df.info()
```

```python
plt.figure(figsize=(6, 6))
plt.title('Year.Exp vs Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(df.YearsExperience,df.Salary,color = "red", marker = '+')
plt.show()
```

```python
x = df[['YearsExperience']]
y = df['Salary']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2,random_state = 10)
```

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
lr_model =lr.predict(X_test)
print(lr_model)
```

```python
accuracy = lr.score(X_test,Y_test)

print("accuracyof the model:",accuracy)
```

```python
predict = lr.predict([[1.2]])

print("predicted value :", predict)
```

```python
plt.figure(figsize=(6, 6))
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.scatter(Y_test, lr_model, color='blue', edgecolor='k')
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='red', linestyle='--')  # y=x line
plt.grid(True)
plt.show()
```
