```python
import pandas as pd
import warnings
warnings.simplefilter("ignore")
from sklearn.linear_model import LogisticRegression , LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
X = df[['Age']]
Y = df['Outcome']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 0)
```

```python
#linear regression
lr_model = LinearRegression()
lr_model.fit(X_train,Y_train)
pred = lr_model.predict(X_test)
```

```python
plt.xlabel('Age')
plt.ylabel('Outcome')
plt.scatter(df.Age, df.Outcome, color='blue', edgecolor='k')
plt.plot(X_test, pred, color='red',linestyle= '--')
plt.show()
```

```python
x = df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y = df['Outcome']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 0)
```

```python
lor_model = LogisticRegression() 
lor_model.fit(x_train,y_train)
```

```python
accuracy = lor_model.score(x_test,y_test)
```

```python
accuracy
```

```python
x_pred = lor_model.predict(x_test)
x_pred
```

```python
evaluate_score = lor_model.predict([[6,148,72,35,0,33.6,0.627,50]])
print("actual value:",y[0])
print("predicted value:",evaluate_score)
```
